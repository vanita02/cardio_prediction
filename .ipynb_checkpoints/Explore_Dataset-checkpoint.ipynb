```python
# histrogram

columns = ['gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']

for col in columns:
    plt.figure(figsize=(8,4))
    sns.histplot(data[col], kde=True, shrink=0.8)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
target = "cardio"

for col in columns:
    if col != target:  
        plt.figure(figsize=(8,5))
        plt.scatter(data[col], data[target])
        plt.title(f"Scatter Plot: {col} vs {target}")
        plt.xlabel(col)
        plt.ylabel(target)
        plt.tight_layout()
        plt.show()
```

```python
df = data.corr()
df
```

```python
#outliers
q1 = data['calculated_age'].quantile(0.25)
q3 = data['calculated_age'].quantile(0.75)
print(q1,q3)

IQR = q3-q1

lower = q1-1.5*IQR
upper = q3+1.5*IQR

data = data[(data['calculated_age']>lower)|(data['calculated_age']<upper)]
data
```

```python
q1 = data['ap_hi'].quantile(0.25)
q3 = data['ap_hi'].quantile(0.75)
print(q1,q3)

IQR = q3-q1

lower = q1-1.5*IQR
upper = q3+1.5*IQR

data = data[(data['ap_hi']>lower)|(data['ap_hi']<upper)]
data
```
